## Neurônios cerebrais

Os neurônios são formados pelos **dendrites** que são responsáveis por *receber as informações de outros neuronios* ou do ambiente externos e armazena no nucleo do neurônio.

Uma vez que tem informação suficiente é captada gerando um impulsos (sinais elétricos), *o somatorio dos impulsos elétricos que chegam no neurônio* passar de um certo limiar de volts, *essa informação é enviada por meio de sinais elétricos*, o **neurônio é ativado**. 

Toda energia que recebeu passa a adiante a sua energiz por meio do **axônio** que seria o *corpo do neurônio*.

Essa energia é entrege para a **sinapse** que transmite para outros neuronios.

>***Funcionamento***: O neurônio recebe uma carga elétrica da sua vizinhança e se a carga eletrica form superior ao certo limiar, ele é atividado e passará sua carga elétrica adiante até a sinapse, lenado até o proximo neuronio ate chegar no cerebro que processa esse impulso.

**Figura 1** - Modelo de Mcculloch-Pitts (1943)



Fonte: Lead Dell

**Neurônio possui uma computação simples**, recebe um sinal e, uma vez que tal sinal seja suficiente será passado esse sinal adiante. Nessa maneira pequenas partes irão influenciar no todo, um neuronio com computação simples mas a junção de diversos neuronios tras mais ganha. 

**Camadas de neurônios**:
Em uma situação em que seja necessário o uso de várias camadas, as primeiras camadas iram detectar as nuncces mais leves do que as próximas camadas. As camadas do meio iram receber as informações da camadas iniciais, que detectou sinais mais simples, serão agregados a sinais mais complexos nessa camadas mais internas e por fim será agrupado esses sinais de uma maneira ainda mais complexa e, assim, ***as camadas vão especializando os dados***, ou seja, especializando o impulso.

>Dessa maneira, temos camadas de neurônios onde cada uma é formada por diversos neurônios e a medida que o sinal vai ser processado o pulso elétrico é processado de uma camada a outra temos a evolução do precessamento e a especializando o modelo, tornando o sinal mais complexos.

## Neurônio artifical

O neuronio artifical possui as seguintes caracterisiticas:

- Conjunto de variaveis de entradas:
Cada uma dessas váriaveis carrega uma informação diferente. Assim, cada variável representa um impulso externo, sendo traduzidos comos dados de entrada, ou o resultado da computação de um outro neurônio anterior.

- Processamento do neurônio: 
Dentro do neurônio as informações de entrada são processadas.

- Saída do neuronio
Resposta do neurônia as entradas.

### Perceptron

Parecido com imagem anterior, já que possui o conceito de somatório das entradas, porém existe agora um peso que multiplica cada uma das variaveis de entradas e que é independente dadas demais.

Dessa maneira, o neurônio da uma diferenciada nas informações que estão chegando nos neuronios, assim é possível diferencias a entrada tornando um neurônio mais sensível a um entrada em relação a outra. Assim, é possivel ter uma complexidade maior na informação que está chegando nesse neurônio.

### Função de ativação

Um vez que essa somas das entradas são calculadas , ou seja, soma-se os oprodutos de cada variável pelo seu peso, esse somatório será passado por uma função ativação que tem a **função** de tranformar o valor os valores do samatório em um valor determinado.

Na Figura 4, a função ativação retorna 1 ou retorna 0.

Tipos de função:
- Função linear: entra  exatamente o valor do somatório
- ReLU (Rectified Linear Unit): Apenas retifica os valore menores que zero, ou seja, se o somatorio do processamento for igual a zero ela retornará zero, caso contrátio, retornará um valor correspondente. Usada em reconhecimento de imagem e padroes.

## Multilayar percetron


Percetron de multicamadas são os neurônios conectados entre si, assim temos os neuronios conectados as entradas de  e as camadas seguintes irão receber como entrada as saídas das camadas anterios de neurônios. Dessa maneira, os **neurônios são especializados**.

O exemplo do ou-exclusivo tornou os dados linearmente não separados, ou seja, não é possível separar os dados utiizando uma reta. *A rede neural pode ser não linear*

**Figura 5** - Porta lógica "ou-exclusivo"



Fonte: Lead Dell

A solução utilizadando esse problema é utilizar dois neurônios, onde a entrada X1 entra no neurônio 1 com um peso e entra no neurônio 2 com outro peso. Nessa situação, teremos 4 pesos diferentes.

**Figura 6** - Rede neural com dois neurônios



Fonte: Lead Dell

Como os neurônios são formados pelo somatorio dos sinais de entrada multiplixado por um peso, temos que os **neurônios podem ser representados com uma reta**. O **coeficiente angular da reta** é um incremento ou descremento em cada neurônio chamado de ***bias***

**Figura 7** - Representaçaõ dos neurônios com retas



Fonte: Lead Dell

Utilizando mais uma cada de neurônio é possível receber o resultado dessas duas primeiras retas, uma que nega o "OU" e outra reta que separa os dois valores de entrada como "E" e vai torna-los verdadeiros, atribuindo pesos a cada uma deles e contruir uma nova reta que tera a semântica: Caso os valores estejam abaixo das duas retas ele é falso, caso ele esteja acima das duas retas, ele também é falso e, se tiver entre as duas retas, será verdadiro.


**Figura 7** - Rede neural do modelo "OU-EXCLUSIVO"



Fonte: Lead Dell

Assim, quanto mais complexo o modelo, mais neurônios que iremos precisar em cada camada para separar os dados. E a medida que se precisa de combinações *entre as retas anteriores, ou seja, precise de mais neuronios para representa o problema* é possivel utilizar mais camadas, e cada cada se especializa em uma carateristica dos dados.

## Implementação de redes neurais

```python
#Importando a rede neural
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn. datasets import load_boston
```

### Base de dados

```python
X, y = load_boston(return_X_y=True)
```

```python
#Separar a base de dados em treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X,y)
```

Pela maneira de treinar a rede neural, usando o back propagasion, é preciso ter cuidado ao usar o *minmax_scala*.

```python
mm = MinMaxScaler()
X_treino = mm.fit_transform(X_treino)
X_teste = mm.transform(X_teste)
```

### Treinando a rede

Possui por padrão a ativação de ReLU, e uma camada oculta com 100 neuronios. *hidden_layer_sizes* é o tamanho da quantidade de camada ocultas com a quantidade de neurônios ocultos. A quantidade máxima de interação *max_iter* por padrão é 100.

```python
# criando a rede
mlp = MLPRegressor(hidden_layer_sizes=(100,100,100,100,100,100), max_iter=1000)
```

```python
# treinando a rede
mlp.fit(X_treino,y_treino);
```

### Validação da rede neural

```python
mlp.score(X_teste,y_teste)
```
