# Tradeoff viés \ Variância

Se as previsões correnspodema aos resultados esperados para o contexto do problema, por isso é importante observar se o modelo possui Overfitting ou Underfitting. Em grande verdade, encontrar o equilibrio entre esses dois conceitos no modelo é objetivo.

O radeoff é uma escolha qu epode causar um custo mas que traz beneficios, assim como na expressão "Perder de um lado para ganhar de outro".

> O Tradeoff é dividido nas subcategorias sendo o **viés** e a **variânça.** 

>A composição e quantidade dos dados em machine learning, assim como o tipo de complexidade dos modelos propostos para resolver um problema, pode afetar a qualidade das previsões, podendo causar tanto o Underfitting, quanto o Overfitting.

## Viés

O tipo de erro na diferença entre os valores da base de dados e os valores previsto, asim o modelo é dito **viesado** quando não aprende corretamente ***fazendo previsões incorretas***. Quando o **viés é baixo** os devios entre as observações e as previsões dimininuem fazendo com que o modelo seja mais preciso.

## Variância


A **Variânca** é uma medida de dispersão que mede o grua de variação dos valores  em relação a sua média, mostrando o quão distante os dados estão do produto central em um conjunto de dados. 

Quando os dados forem treinas e "memorizados" pela ML ao ser submetido a novos dados muitos diferentes na etapa de teste, as distancias entre os pontos do modelo aumentará acasionando uma alta variabilidade ou variância.

Quando **o modelo se especializa nos dados de treino** mas não é capaz de fazer generalizações, ou seja, o modelo não consegue prever corretamente quando surgem dados que ele não conheça.

>Assim modelos que apresentam variança alto são modelos com Overfitting.

## Tradeoff

O ponto em que é escolhido o modelo baseado na variancia e no viés é chamado de Tradeoff viés \ Variância. A escolha se torna difil, já que se diminuir a complexidade do modelo, pode ocrrer Underfitting e se aumentada pode ocorrer Overfitting. 

Existe algumas técnicas para melhor o tradeoff.

### Redução na dimensionalidade dos dados

Constitui na tentativa de descrever os dados por meio da maioria dos atributos exceto os atributos que tenham pouca ou nenhuma relação com o atributo de saída. Além de reduzir a qualidades de atributos desncessários. 

### Validação cruzada

O conjunto de dados é submetido ao treino e ao teste várias vezes, porém usando uma parte diferente dos dados para cada rodada de treino e teste. Desse modo, como geralmente utiliza-se 20% a 30% dos dados de treino e o restante é para teste.
