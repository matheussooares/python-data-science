# Classificação

É um dos tipos de métodos de *aprendizados supervisionados*.

**Definição**: Objetivo de classificar ou categorizar uma informação com base no conjunto de informações previamente conhecidas.



## Tipos de classificação

- **Classificação Binária**: Há apenas dois tipos de classes de resposta, ou seja duas opções de escolha. Exemplo: se um tive vai ganhar ou não uma partida (vitória ou derrota). 


- **Classificação Multiclasse**: Classifica a resposta de interesse em duas categorias ou mais classe.


## Modelo de classificador

Possui duas etapas, a de treino e a de teste de algoritmo. A etapa de treino do algoritmo consiste em passar os dados de entrada e por meio de interações será buscado padrões de similiridade. A etapa de teste do algoritmo pega dados não treinados para analisar se o modelo está acertando, com o método de validação.
