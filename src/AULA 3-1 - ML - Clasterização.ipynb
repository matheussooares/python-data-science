# Clusterização

Técnica de ML que faz parte do aprendizado não supervisonado, já que não se conhece os dados previamente , conhecida como análise de agrupamentos. 

Um algoritmo de clustering possui a característica de formar grupos com os itens que possuem características mais semelhantes, que assim são definidas com base em cálculos de distâncias. 

Possui o **bjetivo de agrupar automaticamente um número k de diferentes grupos em um determinado conjunto de dados**, esses grupos são chamados de *clusters* sendo cada um formado por uma parte de dos dados que são mais parecidos.

Dado uma caixa de brinquedo que se deseja organiza-los de acordo com o tipo de cada um, sem saber previamente tipos de os brinquedos dentros da caixa. A primeiras ornanizações acontecem de maneira não definitiva, como visto na figura.  

Para estabelecer a diferença ou a semelhança entre os itens de um conjunto de dados é usado uma **função de distância**. As mais utilizadas são:
- Distância de Manhattan: Simula os pontos que um taxi passa de um ponto a outro passando pelos quarteirões da cidade, considerando *A soma dos modulos das distâncias entre os pontos*.

- Distância de Euclidiana: Utiliza a raiz quadrada da soma das difernças, ao quadrado entre os pontos

- Distância de Minkowski: Generalização  das duas distâncias mencionadas acima.

- Distância de Mahalanobis: Considera a correlações existentes em um conjunto de dados.



## Tipos de algorimos de clusterização

A analise  de agrupamento pode ser dividida em métodos:


### Hierárquicos

Os algoritmos organizam os dados em uma estrutura de árvore binária ou em um dendograma. Cada nó da arvore possui uma informação. 

A arvore binária é composta por nós, raiz e folhas. O dendograma é um tipo de arvore que subdivide os dadis em outros subconjuntos de menor tamanho.

São divididos entre:
- Aglomerativos: Os dados são separados em diversos clusters e são agrupados em clusters maiores até que se tenha um unico clusters.


- Divisivos: Os dados começa juntos em um unico clusters e vão se desmembrando até que se tenha diversos clusters, onde cada um dele representa um dado.




### Não hierárquicos

Dividem os dados em k-clusters de acordo com a distância calculadas entre os dados e o centro de cada clusters. Dessa forma, cada cluster conterá os dados quem possuem a menor distância entre seus elementos e o centro do clusters, ou seja, os dados com caracteristicas semelhantes.
