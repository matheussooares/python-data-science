# Overfitting e Underfitting

Os modelos de marchine learning busca padrão nos dados e a partir disso predizer os resultados de outros dados.

Além disso, a **qualidade das previsões** dos modelos depende de quão bem ele consegue *generalizar suas respostas de previsão* a medida que é submetido a dados não visto no treino.

>Existe dados que prejudicão a previsão de modelos de ML os Overfitting e Underfitting.

# Overfitting

O Overfitting ou sobre-ajuste, ocorre quando o modelo se adapta muito bem ao dadas de treinamentos, porém na etapa de teste na presença de novos dados o modelo não consegue generalizar, apresentando resultados incorretos.

>Quando um modelo de Machine Learning é especializado em reconhecer os dados na etapa de treinamento e não consegue fazer generalizações, ao ser submetido a novos dados, ocorre o que é chamado de Overfitting.

>Os modelos de Machine Learning, que memorizam todo o conjunto de dados de treino, mas não conseguem realizar corretamente as previsões, são caracterizados pelo Overfitting.
